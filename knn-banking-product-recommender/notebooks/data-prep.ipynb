```python
# Convert the features into their intuitive types
ds_train['fecha_dato'] = pd.to_datetime(ds_train['fecha_dato'])
ds_train.age = pd.to_numeric(ds_train.age, errors='coerce')
ds_train.gross_income = pd.to_numeric(ds_train.gross_income, errors='coerce')
ds_train.cust_seniority = pd.to_numeric(ds_train.cust_seniority, errors='coerce')
ds_train.cust_start_date_first_holder_contract = pd.to_datetime(ds_train.cust_start_date_first_holder_contract, errors = 'coerce')

# test dataset
ds_test['fecha_dato'] = pd.to_datetime(ds_test['fecha_dato'])
ds_test.age = pd.to_numeric(ds_test.age, errors='coerce')
ds_test.gross_income = pd.to_numeric(ds_test.gross_income, errors='coerce')
ds_test.cust_seniority = pd.to_numeric(ds_test.cust_seniority, errors='coerce')
ds_test.cust_start_date_first_holder_contract = pd.to_datetime(ds_test.cust_start_date_first_holder_contract, errors = 'coerce')
```

```python
ds_train.dtypes
```

```python
# Data Imputation

# For features with missing values percentage < 10%, use the most common factor level to impute
cols = ['emp_index','cust_country_res','sex','cust_start_date_first_holder_contract','new_cust_index','cust_primary_type',"cust_type_at_start_month","cust_rel_type_at_start_month",
       "province","province_name","activity_index","channel_joined","cust_category"]

for i in cols:
    ds_train.loc[ds_train[i].isnull(), i] = ds_train[i].value_counts().index[0]
    ds_test.loc[ds_test[i].isnull(), i] = ds_test[i].value_counts().index[0]

# For features with missing value accounts for over 10%, impute their missing values based on the mean at different dimensions
# For Income, impute the missings using the medians of Customer Segment and Province
ds_train['gross_income'].fillna(ds_train['gross_income'].mean(), inplace = True)
ds_test['gross_income'].fillna(ds_test['gross_income'].mean(), inplace = True)
```

```python
# Drop columns
# Drop cust_last_primary_date (ult_fec_cli_1t) and spouse_index (conyuemp) as these cols have > 99% missing values

ds_train.drop(['cust_last_primary_date','spouse_index'], axis = 1, inplace = True)
ds_test.drop(['cust_last_primary_date','spouse_index'], axis = 1, inplace = True)
```

### Feature Engineering 

```python
# Group the age for lifestages: young, adult, senior and elder
age_group = [0, 20, 45, 65, 100]
age_labels = ['young', 'adult', 'senior', 'elder']

ds_train['age_grouped'] = pd.cut(ds_train.age, bins = age_group, labels = age_labels)
ds_test['agg_grouped'] = pd.cut(ds_test.age, bins = age_group, labels = age_labels)
```
