```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

## Config

```python
# HF_REPO_ID = "Helsinki-NLP/opus-mt-tc-big-en-fr"
HF_REPO_ID = "Helsinki-NLP/opus-mt-en-fr"
LOCAL_SAVE_DIR = "../saved_model"
MODEL_KWARGS = {
    "do_sample":True,
    "max_new_tokens": 40,
    "top_k":20,
    "top_p":0.95,
    "temperature":0.5
}
```

## Load model

```python
# Load from Hugging Face
tokenizer = AutoTokenizer.from_pretrained(HF_REPO_ID)
model = AutoModelForSeq2SeqLM.from_pretrained(HF_REPO_ID)

# Save to disk
tokenizer.save_pretrained(LOCAL_SAVE_DIR)
model.save_pretrained(LOCAL_SAVE_DIR)
```

```python
# Load from disk
tokenizer = AutoTokenizer.from_pretrained(LOCAL_SAVE_DIR)
model = AutoModelForSeq2SeqLM.from_pretrained(LOCAL_SAVE_DIR)
```

## Get prediction

```python
text = "My name is Sarah and I live in London"
# text = "This is me"
```

```python
# Get translation
inputs = tokenizer(text, return_tensors="pt").input_ids
outputs = model.generate(inputs, **MODEL_KWARGS)
translation = tokenizer.decode(outputs[0], skip_special_tokens=True)
translation
```
